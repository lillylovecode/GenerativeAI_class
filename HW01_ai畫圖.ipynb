## 畫四葉草圖形

1.先旋轉愛心

```python
# Generate points for the heart shape
t = np.linspace(0, 2*np.pi, 100)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# Rotate the points by 90 degrees
x_rotated, y_rotated = y, x  # Swap x and y for 90-degree rotation

# Fill the rotated heart
plt.fill(x_rotated, y_rotated, color='red')

# Remove axes
plt.axis('off')

# Display the plot
plt.show()
```

2.再旋轉90度(共180度)

```python
# Generate points for the heart shape
t = np.linspace(0, 2*np.pi, 100)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# Rotate the points by 180 degrees
x_rotated = -x
y_rotated = -y

# Fill the rotated heart
plt.fill(x_rotated, y_rotated, color='red')

# Remove axes
plt.axis('off')
```

3.270度=-90度

```python
# Generate points for the heart shape
t = np.linspace(0, 2*np.pi, 100)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# Rotate the points by 270 degrees (equivalent to -90 degrees)
x_rotated = -y  # Swap and negate x
y_rotated = x   # Swap x and y

# Fill the rotated heart
plt.fill(x_rotated, y_rotated, color='red')

# Remove axes
plt.axis('off')
```

4.合併四個愛心成四葉草

```python
def draw_heart():
    t = np.linspace(0, 2*np.pi, 100)
    x = 16 * np.sin(t)**3
    y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)
    return x, y

# 畫出4個愛心
x1, y1 = draw_heart()

x2, y2 = draw_heart()
# 將每個點旋轉90度
x2, y2 = y2, x2
x2, y2 = x2+20, y2-15 #移動頂點

x3, y3 = draw_heart()
# 將每個點旋轉180度
x3, y3 = -x3, -y3
x3, y3 = x3, y3-30 #移動頂點

x4, y4 = draw_heart()
# 將每個點旋轉270度 (相當於-90度)
x4, y4 = -y4, x4
x4, y4 = x4-20, y4-15 #移動頂點

#將愛心用綠色填滿
plt.fill(x1, y1, color='green')
plt.fill(x2, y2, color='green')
plt.fill(x3, y3, color='green')
plt.fill(x4, y4, color='green')

#關閉座標軸
plt.axis('off')
```
